```python
from selenium import webdriver
import time
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select

web = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
web.get('http://localhost/Mini-PBL/mahasiswa/Form-Pengajuan.php')
web.maximize_window()

username = web.find_element(By.XPATH, '//*[@id="exampleInputEmail1"]')
# username.click()
username.send_keys("mahasiswa")


password = web.find_element(By.XPATH, '//*[@id="exampleInputPassword1"]')
# password.click()
password.send_keys("mahasiswa123")

login = web.find_element(By.XPATH, '/html/body/div[1]/div/div[2]/form/div[4]/button')
login.click()

def peminjaman(nama_anggota,keperluan,i,select_room,select_dosen,tanggal,jm,js):
    form_peminjaman = web.find_element(By.XPATH, '/html/body/div[1]/div[1]/div[2]/div[3]/a/span')
    form_peminjaman.click()

    Nama_nama_Anggota_yang_Ikut = web.find_element(By.XPATH, '/html/body/div[1]/div[2]/div/div[2]/div[2]/form/div[3]/div/div/textarea')
    Nama_nama_Anggota_yang_Ikut.send_keys(nama_anggota)

    Keperluan = web.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div[2]/div[2]/form/div[4]/div/div/textarea')
    Keperluan.send_keys(keperluan)

    Hari = web.find_element(By.XPATH,f'//*[@id="hari"]/option[{i+1}]')
    Hari.click()
    
    Ruangan = web.find_element(By.XPATH, f'//select[@id="ruang"]//option[@value="{select_room}"]') 
    Ruangan.click()
    
    Penanggung_Jawab = web.find_element(By.XPATH, f'//select[@name="dosen_penanggung_jawab"]/option[@value="{select_dosen}"]')
    Penanggung_Jawab.click()

    Tanggal = web.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div[2]/div[2]/form/div[8]/div/div/div/input')
    Tanggal.send_keys(tanggal)

    Jam_mulai = web.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div[2]/div[2]/form/div[9]/div/div/div/input')
    Jam_mulai.send_keys(jm)

    Jam_selesai = web.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div[2]/div[2]/form/div[10]/div/div/div/input')
    Jam_selesai.send_keys(js)

    time.sleep(1)
    submit = web.find_element(By.XPATH,'/html/body/div/div[2]/div/div[2]/div[2]/form/div[11]/div/input')
    submit.click()

    cek = web.find_element(By.XPATH, '/html/body/div[1]/div[2]/div/div[2]')
    return cek.text

for i in range(int(input('Masukkan Jumlah Pinjaman: '))):
    nama_anggota = input('Nama anggota yang ikut: ')
    keperluan = input('Keperluan: ')
    i = int(input('Hari peminjaman:(Senin:1,Selasa:2,....,Minggu:7) '))
    select_room = input('Ruangan yang akan dipinjam : ')
    select_dosen = input('Penanggung Jawab: ')
    tanggal = input('Tanggal peminjaman(Format:MMHHYYYY) ')
    jm = input('Jam mulai(Format:HHMM(A untuk Am atau P untuk PM)) ')
    js = input('Jam selesai(Format:HHMM(A untuk Am atau P untuk PM)) ')
    print(peminjaman(nama_anggota,keperluan,i,select_room,select_dosen,tanggal,jm,js))
    fp = web.find_element(By.XPATH, '/html/body/div[1]/div[1]/div[2]/div[3]')
    fp.click()
    
    
    
```
